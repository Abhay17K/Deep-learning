```python
import numpy as np
import pandas as pd
df = pd.read_csv('/content/diabetes.csv')
df.head(20)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.describe()
```

```python
df['Glucose'] = np.where(df['Glucose']==0, df['Glucose'].mean(), df['Glucose'])
df['BloodPressure'] = np.where(df['BloodPressure']==0, df['BloodPressure'].mean(), df['BloodPressure'])
df['SkinThickness'] = np.where(df['SkinThickness']==0, df['SkinThickness'].mean(), df['SkinThickness'])
df['Insulin'] = np.where(df['Insulin']==0, df['Insulin'].mean(), df['Insulin'])
df['BMI'] = np.where(df['BMI']==0, df['BMI'].mean(), df['BMI'])
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
y
```

```python
# Mandatory preprocessing - Data Normalization ***pls remind me tomorrow regarding data normalization
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
```

```python
# split the data into training and test
from sklearn.model_selection import train_test_split
x_train, x_test, y_train , y_test = train_test_split(x, y, test_size=0.2, random_state=101)
```

# MultiLayer Perceptron Method - DNN/ANN method

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from keras.layers import Dense
```

```python
dnn = Sequential()
dnn.add(Dense(32, activation='relu', input_dim=8))
dnn.add(Dense(32, activation='relu'))
dnn.add(Dense(1, activation='sigmoid'))
dnn.summary()
```

```python
dnn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
dnn.fit(x_train, y_train, validation_data=(x_test, y_test), epochs=100)
```
